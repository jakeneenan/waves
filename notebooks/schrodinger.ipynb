```python
import matplotlib.pyplot as plt
import numpy as np
#Constants
h = 6.626e-34
m = 9.11e-31
#Values for L and x
x_list = np.linspace(0,5,100)
L = 5
def psi(n,L,x):
    return np.sqrt(2/L)*np.sin(n*np.pi*x/L)
def psi_2(n,L,x):
    return np.square(psi(n,L,x))
```

```python
plt.plot(x_list, psi(1,L,x_list), label='n=2')
plt.grid()

# Save the plot as a pdf
#plt.savefig('../output/psi.pdf')
```
